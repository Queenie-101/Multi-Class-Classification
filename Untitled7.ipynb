```python
#2. Library importation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#3. load the data
ds=pd.read_csv('unconv_MV_v5.csv')
ds.head()
```

```python
#4. Dataset information
ds.info()
```

```python
#5. check for null values
ds.isnull().sum()
```

```python
#6. feature extraction
x=ds.drop(['Prod'],axis=1) #input features
y=ds['Prod']#Target feature-what to predict
```

```python
#7. Split the data into testing and training
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#8. load the model
model=LinearRegression()
```

```python
#9. train the model
model.fit(train_x,train_y)
```

```python
#10. make predictions with the model
predictions=model.predict(test_x)
```

```python
#11. Compare the predicted values with actual values
pd.DataFrame({'Actual values':test_y, 'Predicted values':predictions})
```

```python
#12. Evaluate the model's accuracy
accuaracy=model.score(test_x,test_y)
print(f'Accuracy of the model is {accuaracy*100:.2f}%')
```

```python
#plot a graph to show the accuracy of the algorithm
sns.scatterplot(x=test_y,y=predictions)
plt.title('Actual vs Predicted')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```
